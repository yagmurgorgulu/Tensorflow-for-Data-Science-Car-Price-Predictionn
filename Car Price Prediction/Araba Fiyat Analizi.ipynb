```python
df = yuzde99df
```

```python
df = df[df.year !=1970]
```

```python
df.groupby("year").mean()["price"]
```

```python
df = df.drop("transmission" , axis=1)
```

```python
df.head()
```

```python
y = df["price"].values
x = df.drop("price",axis=1).values #y=ax+b de y ve x leri belirleyip bunları numpy dizisine dönüştürdük.
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi
```

```python
kayipVerisi.plot() #loss ve val_loss değerleri birlikte hareket etmeli eğer bitiş çizgisinde biri aşağı biri yukarı gidiyorsa overfitting durumu var epochs u azaltmam gerek
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi) #fark kabul edilebilir olabilir eğer denemek istrsek yukarıda değiştirerek bakabiliriz.
```

```python
plt.scatter(y_test,tahminDizisi)
plt.plot(y_test,y_test,"g-*")
```
